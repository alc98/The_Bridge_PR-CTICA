Obtener el nombre  de los empleados que hayan repartido al bar Stop durante la semana del 17 al 23 de octubre de 2005. 

```python
cursor.execute('''
SELECT DISTINCT E.Nombre
FROM Empleados E
JOIN Reparto R ON E.CodE = R.CodE
JOIN Bares B ON R.CodB = B.CodB
WHERE B.Nombre = 'Stop'
  AND R.Fecha BETWEEN '10/17/05' AND '10/23/05'
''')
```

Obtener el Cif y nombre de los bares a los que se ha repartido cerveza de tipo Botella y 
# capacidad inferior a 1 litro, ordenados por localidad 

```python
cursor.execute('''
SELECT DISTINCT B.Cif, B.Nombre
FROM Bares B
JOIN Reparto R ON B.CodB = R.CodB
JOIN Envase E ON R.CodC = E.CodC
WHERE E.Tipo = 'Botella'
  AND E.Capacidad < 1
ORDER BY B.Localidad
''')
```

SELECT DISTINCT B.Cif, B.Nombre: seleccionamos el CIF y el nombre de los bares, sin duplicados.

FROM Bares B JOIN Reparto R ON B.CodB = R.CodB JOIN Envase E ON R.CodC = E.CodC: hacemos joins para relacionar los bares con los repartos y los envases que se entregaron.

WHERE E.Tipo = 'Botella' AND E.Capacidad < 1: filtramos solo los repartos de cervezas tipo Botella y con capacidad menor a 1 litro.

ORDER BY B.Localidad: ordenamos los resultados por la localidad del bar para verlos organizados geográficamente.

El resultado será una lista de bares que recibieron botellas pequeñas de cerveza, mostrando su CIF y nombre, ordenados por localidad.

bares a los que se les ha repartido envases de tipo botella y capacidad 0.2 ó 0.33 

```python

cursor.execute('''
SELECT DISTINCT B.Nombre AS Bar
FROM Reparto R
JOIN Bares B ON R.CodB = B.CodB
JOIN Envase E ON R.CodC = E.CodC
WHERE E.Tipo = 'Botella'
  AND E.Capacidad IN (0.2, 0.33)
''')
```

Nombre de los empleados que han repartido a los bares "Stop" y "Las Vegas" cervezas con envase botella. 

```python
cursor.execute('''
SELECT DISTINCT E.Nombre
FROM Empleados E
JOIN Reparto R ON E.CodE = R.CodE
JOIN Bares B ON R.CodB = B.CodB
JOIN Envase C ON R.CodC = C.CodC
WHERE B.Nombre IN ('Stop', 'Las Vegas')
  AND C.Tipo = 'Botella'
''')
```

Obtener el nombre y número de viajes que ha realizado cada empleado fuera de Villa Botijo. 

```python
cursor.execute('''
SELECT E.Nombre, COUNT(*) AS NumViajes
FROM Empleados E
JOIN Reparto R ON E.CodE = R.CodE
JOIN Bares B ON R.CodB = B.CodB
WHERE B.Localidad != 'Villa Botijo'
GROUP BY E.Nombre
''')
```

Obtener el nombre y localidad del bar que más litros de cerveza ha comprado.  

```python
cursor.execute('''
SELECT B.Nombre AS Bar, B.Localidad, SUM(E.Capacidad * R.Cantidad) AS Litros
FROM Reparto R
JOIN Bares B ON R.CodB = B.CodB
JOIN Envase E ON R.CodC = E.CodC
GROUP BY B.CodB
ORDER BY Litros DESC
LIMIT 1
''')
```

Obtener los bares que han adquirido todos los tipos de cerveza con envase de botella y capacidad menor que 1 litro. 

```python
cursor.execute('''
SELECT B.Nombre
FROM Bares B
WHERE NOT EXISTS (
    SELECT *
    FROM Envase E
    WHERE E.Tipo = 'Botella' AND E.Capacidad < 1
      AND NOT EXISTS (
          SELECT *
          FROM Reparto R
          WHERE R.CodB = B.CodB AND R.CodC = E.CodC
      )
)
''')
```

```python
cursor.execute('''
SELECT E.CodE
FROM Empleados E
JOIN Reparto R ON E.CodE = R.CodE
GROUP BY E.CodE
ORDER BY COUNT(DISTINCT R.Fecha) DESC
LIMIT 1
''')
```
