3. Sabemos que podemos pintar gr치ficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
personas = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']

alturas = [160, 180, 175, 190, 170]
```

```python
plt.bar(personas, alturas, color='green')

plt.xlabel('Personas')
plt.ylabel('Altura (cm)')
plt.title('Altura de un grupo de amigos')

plt.show()
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
barplot = plt.bar(personas, alturas, color='green')

for bar in barplot:
    altura = bar.get_height()          
    x_pos = bar.get_x() + bar.get_width()/2  # centro de la barra
    plt.text(x_pos, altura, str(altura), ha='center', va='bottom')
```

```python
plt.xlabel('Personas')
plt.ylabel('Altura (cm)')
plt.title('Altura de un grupo de amigos')

plt.show()
```

22. Contornos de 3D a 2D. Escribe una funci칩n que recibiendo x,y devuelva (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

Tablero: Tanto x como y van de -3 a 3 y usaremos 256 puntos.
Usa contourf, con 8 niveles (cortes), una transparencia de 0.75 y un color map de tipo 'jet'. 
Bonus: pinta las l칤neas de los contornos tambi칠n de negro

```python
def f(x, y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 - y**2)
```

```python
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)  # crear malla 2D
```

```python
Z = f(X, Y)
```

```python
plt.figure(figsize=(6, 5))
contours = plt.contourf(X, Y, Z, levels=8, alpha=0.75, cmap='jet') 
```

```python
plt.contour(X, Y, Z, levels=8, colors='black', linewidths=0.5)

plt.title('Contornos 2D de la funci칩n f(x,y)')
plt.xlabel('x')
plt.ylabel('y')

plt.colorbar(contours) 
plt.show()
```
