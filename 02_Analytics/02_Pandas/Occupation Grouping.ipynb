```python
users['gender_n'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)


ratio_per_occ = users.groupby('occupation')['gender_n'].mean()

# Asignar el ratio correspondiente a cada usuario según su ocupación
users['ratio'] = users['occupation'].map(ratio_per_occ)

users = users.sort_index()

print(users)
```

```python
# Calcular cuántas personas hay por ocupación y género
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)

# Calcular los porcentajes (ratio)
gender_ratio = gender_counts.div(gender_counts.sum(axis=1), axis=0) * 100


ratio_long = gender_ratio.stack().reset_index()
ratio_long.columns = ['occupation', 'gender', 'ratio']

print(ratio_long)
```

```python
male_count = users.groupby('occupation')['gender_n'].sum()

# Calcular el porcentaje respecto al total de hombres
male_percentage = (male_count / male_count.sum()) * 100

male_percentage = male_percentage.sort_values(ascending=False)

print(male_percentage.round(2))
```

```python
male_count = users.groupby('occupation')['gender_n'].sum()

male_percentage = (male_count / male_count.sum()) * 100

male_percentage = male_percentage.sort_values(ascending=False)

male_percentage = male_percentage.map(lambda x: f"{x:.2f}%")

print(male_percentage)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
age_stats = users.groupby('occupation')['age'].agg(['min', 'max'])

print(age_stats)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age = users.groupby(['occupation', 'gender'])['age'].mean().reset_index()

gender_count = users.groupby(['occupation', 'gender']).size().reset_index(name='count')

# Combinar edad media con el conteo
combined = mean_age.merge(gender_count, on=['occupation', 'gender'])

# Seleccionar el género mayoritario por ocupación
majority_gender = combined.loc[combined.groupby('occupation')['count'].idxmax()]

# Seleccionar y renombrar columnas
result = majority_gender[['occupation', 'gender', 'age']].rename(columns={
    'occupation': 'Profesion',
    'gender': 'Genero Mayoritario',
    'age': 'Edad Media'
})

result['Edad Media'] = result['Edad Media'].round(2)

print(result)
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_count = users.groupby(['occupation', 'gender']).size()

gender_count['Total'] = gender_count.sum(axis=1)

# Calcular porcentaje de cada género
gender_count['%Hombres'] = (gender_count.get('M', 0) / gender_count['Total'] * 100).round(2)
gender_count['%Mujeres'] = (gender_count.get('F', 0) / gender_count['Total'] * 100).round(2)

# Determinar el género mayoritario
gender_count['Genero Mayoritario'] = gender_count.apply(lambda row: 'Hombres' if row.get('M',0) >= row.get('F',0) else 'Mujeres', axis=1)

result = gender_count[['%Hombres', '%Mujeres', 'Genero Mayoritario']].reset_index()

print(result)
```
